```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as st
import plotly.graph_objects as go
```

```python
data = pd.read_csv('AmesHousing.csv')
```

```python
data["SalePrice"].describe()
```

```python
#make a correlation matrix to see what features are important to sale price

corr_matrix = data.corr()
f, ax = plt.subplots(figsize = (12,9))
sns.heatmap(corr_matrix, vmax=.8, square=True)
plt.show()
```

```python
#missing values 

#getting the count of missing values from each column
missing = data.isnull().sum()
#filtering out features with no missing values
missing = missing[missing > 0]
#sorting based on most missing 
missing = missing.sort_values()

missing.plot.bar(title = 'Missing Values by Count')
```

```python
#lets split up the features based on the different data types 

categorical = data.select_dtypes(include="object")
numeric = data.select_dtypes(include='number')

print(len(data.columns))
print(len(categorical.columns))
print(len(numeric.columns))
```

```python
#lets check the distribution of the target value, SalePrice

ax = sns.displot(data = data.SalePrice, kde=True)
plt.xticks(rotation=45)
plt.title("SalePrice Distribution")
plt.show()
```

```python
#Lets take a look at the top 10 features with the highest correlation to the target, saleprice. 

corrSalePrice = data.corrwith(data['SalePrice'])
corrSalePrice.sort_values(inplace=True, ascending=False)

#dropping PID and SalePrice. PID is not needed as it is not meant to correlate to SalePrice
corrSalePrice = corrSalePrice[1:-1]
corrSalePrice[0:9]
```

```python
corrSalePrice.sort_values(inplace=True, ascending=True)
plt.figure(figsize=(14,10),dpi=80)

plt.hlines(y=corrSalePrice.index, xmin=0, xmax=corrSalePrice.values, linewidth=5)
plt.gca().set(xlabel = 'correlation', ylabel = "Features")
plt.title('Correlation of Features to Target SalePrice')  
plt.grid(linestyle = '--', alpha=0.5) 


plt.show()

```

```python
sns.boxplot( x = 'Overall Qual', y = 'SalePrice', data = data).set(title = "Overall Qual Correlation to SalePrice")
```

```python
#next lets look into GrLivArea 

data.plot.scatter(y='SalePrice',x='Gr Liv Area').set(title = "Gr Liv Area scatter plot")
```

```python
#Next lets look at GarageCars. Since this is a range from 0-4 we will use a boxplot instead of a scatter plot. 

sns.boxplot(x='Garage Cars', y="SalePrice", data = data).set(title="GarageCars correlation to sale price")
```

```python
data.plot.scatter(y='SalePrice',x='Garage Area').set(title = "Garage area scatter plot")
```

```python

fig, ax = plt.subplots(figsize=(11.7,5))

NeighborhoodOrder = data.groupby(['Neighborhood']).median().SalePrice.sort_values(ascending = False).index

sns.boxplot(x='Neighborhood', y="SalePrice", data = data, order = NeighborhoodOrder).set(title="Neighborhood Sale Prices")
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
plt.show()
```

```python
#add total sqr feet to list 

data['TotalSqFt'] = data['Total Bsmt SF'] + data['1st Flr SF'] + data['2nd Flr SF']
corrSalePrice = data.corrwith(data['SalePrice'])
corrSalePrice.sort_values(inplace=True, ascending=True)
corrSalePrice = corrSalePrice[1:-1]
plt.figure(figsize=(14,10),dpi=80)

PltColors = ['blue','blue', 'blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue', 'blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','blue','red','blue']

plt.hlines(y=corrSalePrice.index, xmin=0, xmax=corrSalePrice.values, linewidth=5, colors = PltColors)
plt.gca().set(xlabel = 'correlation', ylabel = "Features")
plt.title('Correlation of Features to Target SalePrice')  
plt.grid(linestyle = '--', alpha=0.5) 


plt.show()
```
