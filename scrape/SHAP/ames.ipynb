```python
import pandas as pd 
import numpy as np 
import shap
from sklearn import ensemble

from sklearn.preprocessing import LabelEncoder, OneHotEncoder



```

```python
train = pd.read_csv('C:/Users/Griffin/Documents/GitHub/DS_MRP/Data/amesTrain.csv')
```

```python
train
```

```python

temp = train['MSZoning']

LE = LabelEncoder()
IE = LE.fit_transform(temp)

IE = IE.reshape(len(IE), 1)

OHE = OneHotEncoder(sparse_output=False)
OHE = OHE.fit_transform(IE)

print(OHE)
```

```python
one_hot = pd.get_dummies(train['MSZoning'])

train = train.drop('MSZoning', axis = 1)
train = train.join(one_hot)

one_hot
```

```python
#Sale price set as target value

trainX = train[train.columns[:-1]]
trainY = train[train.columns[-1:]]

regressor = ensemble.RandomForestRegressor()
regressor.fit(trainX,trainY)

explainer = shap.TreeExplainer(regressor)

shap_values = explainer.shap_values(trainX)



```
